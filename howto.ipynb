```python
!pip install -q -U google-generativeai
import google.generativeai as genai
from PIL import Image
GOOGLE_API_KEY = ""
genai.configure(api_key=GOOGLE_API_KEY)
model = genai.GenerativeModel('gemini-2.5-flash')
image_path = '/content/사진'
def ask_ai_tutor(image_path, user_question):
    img = Image.open(image_path)
    prompt = f"""
    당신은 LG전자의 최신 세탁기 전문가 'AI 튜터'입니다.    사용자가 보내준 사진을 보고 질문에 대해 친절하고 정확하게 답변해 주세요.
    
    [규칙]
    0. 사진에 보이지 않는 기능은 절대 추천하지 마세요. 모르면 모른다고 솔직히 말하고 한계를 인정하세요. 
    1. 사진에 있는 버튼이나 옷감을 정확히 식별하세요.
    2. 초보자도 이해하기 쉽게 설명하세요.
    3. 필요하다면 주의사항(경고)도 함께 알려주세요.
    4. 사용자가 잘 사용하지 않는데 지금 상황에 맞는 기능이 있다면 그 기능도 추천해주세요.
    5. 사진의 화질이 좋지 않아서 글자를 인식할 수 없는 경우에는 사진을 다시 제공해달라고 요청하세요.
    
    사용자 질문: {user_question}
    """# API 호출
    response = model.generate_content([prompt, img])
    return response.text

# --- 테스트 실행 영역 ---
```

```python
# 사용자가 조작부 사진을 찍어서 보냈다고 가정
image_path = "/content/사진/세탁기_조작부.jpeg" 
question = "여기서 '스팀' 기능은 어디 있고, 언제 쓰는 거야?"
answer = ask_ai_tutor(image_path, question)
print(answer)
# 예상 답변 결과:# "사진의 오른쪽 상단에 구름 모양 아이콘과 함께 'Steam'이라고 적힌 버튼이 있습니다.# 이 기능은 알러지 케어나 찌든 때를 불릴 때 사용하면 효과적입니다. 3초간 꾹 눌러보세요."
```

```python
# 사용자가 니트와 청바지를 섞어놓은 사진을 보냈다고 가정
image_path = "/content/사진/수건사진.jpeg"
question = "이 옷들 지금 한꺼번에 표준 코스로 돌려도 돼?"
answer = ask_ai_tutor(image_path, question)
print(answer)
# 예상 답변 결과:# "잠시만요! 사진을 보니 두꺼운 청바지와 섬세한 니트 소재가 섞여 있습니다.# 표준 코스로 같이 세탁하면 니트가 줄어들거나 보풀이 생길 수 있어요.# 번거롭더라도 분리 세탁을 추천하며, 니트는 반드시 '울/섬세' 코스를 이용해 주세요."
```
